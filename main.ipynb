### Hypothesis:

- People who smoke will have higher medical costs than those who don't however, there are many more non-smokers in the dataset than smokers so I will be comparing the distribution of medical costs for both groups. 
- People who have children will be more likely to not smoke. 
- Older people will have higher medical costs than younger people.
- Higher BMI will correlate with higher medical costs.
- Males will be the majority of the smokers and will have higher medical costs than females.
- Higher BMI individuals have more children. 
- People from Northeast will have higher medical costs than those from other regions.

#### Checking the first hypothesis

```python
smoke = df["smoker"]
smoker_charges = df[smoke == "yes"]["charges"]
non_smoker_charges = df[smoke == "no"]["charges"]

fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize = (12,5))


# plotting smoker charges distribution 
ax1.hist(smoker_charges, bins=20, color='blue', alpha=0.7)
ax1.set_title("Smoker Charges Distribution")
ax1.set_xlabel("Charges")
ax1.set_ylabel("Frequency")

# plotting non-smoker charges distribution 
ax2.hist(non_smoker_charges, bins=20, color='green', alpha=0.7)
ax2.set_title("Non-Smoker Charges Distribution")
ax2.set_xlabel("Charges")
ax2.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

The smoker charges is a bimodal distribution with a peak around 20,000 and another around 40,000. The non-smoker charges is a right-skewed distribution with a peak around 7,500. This shows that smokers do tend to have higher medical costs than non-smokers.
